```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data
```

```python
data['Quantity']=1
data
```

```python
#transactions = data.groupby(['Member_number', 'itemDescription']).sum().unstack().reset_index().set_index('Member_number')
transactions = data.groupby(['Member_number', 'itemDescription']).size().unstack().reset_index().set_index('Member_number')
```

```python
transactions=transactions.fillna(0)
transactions
```

```python
def encode(x):
    if x<=0:
        return 0
    elif x>=0:
        return 1
basket = transactions.map(encode)
```

```python
basket
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemset = apriori(basket, min_support=0.06, use_colnames=True)
num_itemsets = frequent_itemset
rules = association_rules(frequent_itemset, num_itemsets, metric='lift', min_threshold=1)
```
